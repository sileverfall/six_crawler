## Parsing

```python
next_page = [i.select('td.link > div > table > tr > td > a') for i in soup.select('body > div > table:nth-child(3)')][0][2]['href']
```

```python
last_page = [i.select('td.link > div > table > tr > td > a') for i in soup.select('body > div > table:nth-child(3)')][0][3]['href']
```

```python
%%time
Doctors=list()
page = 0
while next_page != last_page:
    page +=1
    url = url+'&page='+str(page)
    res = requests.get(url,headers = headers)
    res.encoding = 'Big5'
    soup = bs(res.text,"lxml")
    doctors = [i.text for i in soup.select('table.download > div > tr > td')]
    name = doctors[::5]
    clinic = doctors[2::5]
    address = doctors[3::5]
    tel = doctors[4::5]
    doctor=dict()
    for i in range(len(name)):
        doctor={
            'name':name[i],
            'clinic':clinic[i],
            'address':address[i],
            'tel':tel[i]
        }
        Doctors.append(doctor)        
```
